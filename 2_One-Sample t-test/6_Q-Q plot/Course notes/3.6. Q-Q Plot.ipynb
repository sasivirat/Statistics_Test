```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
df = pd.read_excel("3. Case 2 - One-sample t-test.xlsx")
```

```python
df.head()
```

```python
# Plot a histogram of the 'Price_per_EQ' variable

# Set the figure size
plt.figure(figsize=(10, 5))

# Plot the histogram with 30 bins
df["Price_per_EQ"].hist(bins=30)

# Add a title to the plot
plt.title('Price_per_EQ')

# Add the yaxis label
plt.ylabel("Euros")

# Show the plot
plt.show()
```

```python
# Create Q-Q plot for the 1st data set
fig, ax = plt.subplots(figsize=(6,6))
sm.qqplot(df["Price_per_EQ"], line='s', ax=ax, fmt='b')

# Set the chart title and axis labels
ax.set_title("Q-Q plot")

# Display the chart
plt.show()
```

As the dots on the Q-Q plot form a line that is quite straight, we can conclude that the data are normally distributed.
